## Imports

```python
import pandas as pd
import numpy as np
```

## Custom funtions

```python
def grade_setter(x: int) -> str:
    """
    This function takes in a number and returns a letter grade.
    """

    if x == 0:
        return "Undefined Grade"

    if x <= 18:
        return "F"
    elif 19 <= x < 38:
        return "E"
    elif 39 <= x < 58:
        return "D"
    elif 59 <= x < 78:
        return "C"
    elif 79 <= x < 93:
        return "B"
    elif x >= 93:
        return "A"
```

## Data Cleaning - Dealing with NaN values

```python
df.isnull().sum()
```

```python
df = df.fillna(0)
```

## Feature Engineering - Making Letter Grade

```python
# Applying function
df["LetterGrade"] = df["FinalGrade"].apply(lambda x: grade_setter(x))
```
